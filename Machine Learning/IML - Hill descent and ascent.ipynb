## Hill descent algo is used when we want to find the minima of a convex curve and hill ascent algo is used when we want to find maxima of a concave curve. We also use learning rate to avoid over/under shooting

### Hill descent algorithm for 4*x^2 - 38*x + 98

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (4*x**2-38*x+98)
```

```python
# DIFFERENTIAL VALUE
def df(x):
    return (8*x-38)
```

```python
# plot some possible function points
p = np.arange(-10,20,0.25)
plt.plot(p,f(p),'-')
plt.show()
```

```python
# use hill descent to find minima
import time

# start at random point
x = -10
count = 0
alpha = 0.2
# close to 0
tol = 1E-13

# adjust x till convergance
while np.abs(df(x)) > tol:
    count += 1
    print("df(x):", df(x), "x:", x)
    x = x - alpha * df(x)
    time.sleep(0.5)
    

print("x:",x)
print("iterations:",count)
```

### Thus minima is at x = 4.75 and it converged in 68 iterations

## Hill Ascent Algorithm for -4*x^2 + 38 * x - 98

```python
def f1(x):
    return (-4*x**2+38*x-98)

p = np.arange(-15,25,0.25)
plt.plot(p,f1(p),'-')
plt.show()
```

```python
# differential value
def df1(x):
    return (-8*x+38)
```

```python
# use hill ascent to find maxima
import time

# start at random point
x = -15
count = 0
alpha = 0.2
tol = 1E-13

# adjust x till convergance
while np.abs(df1(x)) > tol:
    count +=1
    print("df(x):", df(x), "x:", x)
    x = x + alpha*df1(x)
    #time.sleep(0.5)
    

print("x:",x)
print("iterations:",count)
```

### Thus maxima is at x = 4.75 and converged in 69 iterations
