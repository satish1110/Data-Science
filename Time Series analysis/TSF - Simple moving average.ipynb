### Simple moving average is mainly used to remove random variations and noise from the data. Since observations closer in time are likely to have similar values, these averages are calculted over a sequential set of points in a window length. 

```python
import os 
import pandas as pd
```

```python
os.chdir("C:\\Users\\satish\\Desktop")
```

```python
data = pd.read_excel("TSF_data.xlsx", sheet_name = "SMA")
```

```python
data.head()
```

```python
# plot the data 
data.plot.line(x='date', y='price')
```

```python
# One sided moving average with window length 5 
data['Moving Average'] = data['price'].rolling(window=5).mean()
```

```python
data.head(10)
```

```python
# compare original and sma curve
data.plot.line(x='date',y=['price','Moving Average'])
```

### We can observe above that the random variations of each day have been smoothened out

```python
# Centered moving averages use both previous and future observations to calculate the average at a given point in time.
# Hence centered moving averages use observations that surround it in both directions and these are also called as 
# 2-sided moving averages.

data['Rolling Average Center'] = data['price'].rolling(window=4, center=True).mean()
```

```python
data.head(10)
```

```python
data.plot.line(x='date',y=['price','Rolling Average Center'])
```

### We again observe tht random variations and noise has been reduced. We can also use this to remove seasonal patterns and study underlying trends.
